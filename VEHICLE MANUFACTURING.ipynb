Question:11
Outlier identification:
are there any outlier in the dataset based on price or mileage ?
identifying outlier helps in recognizing unusual patterns that may require further investigation

```python
# Calculate Q1, Q3, and IQR for price
Q1_Price = cars_data['Price'].quantile(0.25)
Q3_Price = cars_data['Price'].quantile(0.75)
IQR_Price = Q3_Price - Q1_Price

# Calculate Q1, Q3, and IQR for mileage
Q1_mileage = cars_data['Mileage'].quantile(0.25)
Q3_mileage = cars_data['Mileage'].quantile(0.75)
IQR_mileage = Q3_mileage - Q1_mileage

# Identify outliers
outliers_Price = (cars_data['Price'] < (Q1_Price - 1.5 * IQR_Price)) | (cars_data['Price'] > (Q3_Price + 1.5 * IQR_Price))
outliers_mileage = (cars_data['Mileage'] < (Q1_mileage - 1.5 * IQR_mileage)) | (cars_data['Mileage'] > (Q3_mileage + 1.5 * IQR_mileage))

# Display outliers
print("Outliers based on Price:")
print(cars_data[outliers_Price])

print("\nOutliers based on Mileage:")
print(cars_data[outliers_mileage])
```

Question 14:
amileage price Relationship:
is there any relationship between mileage and price for specific brands?
answered: investigationg the price between mileage and price aids in pricing strategies


```python


# Assuming 'Brand', 'Mileage', and 'Price' are the columns containing car brand, mileage, and price respectively
# Replace 'Brand', 'Mileage', and 'Price' with the actual column names from your dataset
brands = cars_data['Brand'].unique()

plt.figure(figsize=(12, 8))

for brand in brands:
    brand_data = cars_data[cars_data['Brand'] == brand]
    plt.scatter(brand_data['Mileage'], brand_data['Price'], label=brand, alpha=0.7)

plt.title('Relationship between Mileage and Price for Specific Brands')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.legend()
plt.show()
```

Question 15: Seasonal Price Trend
Can we identify any seasonal trends in the car prices based on the model year?
Answered: analyzing seasonal cars treend helps in understanding market dynamics throughout the year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by model year and calculate mean price for each year
yearly_mean_prices = cars_data.groupby('Model')['Price'].mean()

# Plot the trend
plt.figure(figsize=(10, 6))
yearly_mean_prices.plot(kind='line', marker='o')
plt.title('Average Car Prices Over Model Years')
plt.xlabel('Model Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

Question 16:

```python


# Assuming 'Model_Year' is the column containing model years
# Replace 'Model_Year' with the actual column name from your dataset
model_year_distribution = cars_data['Model'].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(10, 6))
model_year_distribution.plot(kind='bar', color='Blue')
plt.title('Distribution of Cars Based on Model Years')
plt.xlabel('Model Year')
plt.ylabel('Number of Cars')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

question 19: how does the average price of cars change over the years?


```python
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by 'Year' and calculate the average price for each year
average_price_by_year = cars_data.groupby('Year')['Price'].mean()

# Plot the average price over the years
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year.index, average_price_by_year.values, marker='o', linestyle='-')
plt.title('Average Price of Cars Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
